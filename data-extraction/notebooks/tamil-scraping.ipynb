```python
for ix, link in enumerate(links):
    text = ''
    page = requests.get(link, headers={
        'User-rAgent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'
    })
    soup = bs(page.content, 'html.parser')
    
    p_tags = soup.find_all('p')
    
    for tag in p_tags:
        txt = tag.text.strip()
        if txt == '':
            pass
        else:
            if txt[0] in string.ascii_letters:
                pass
            else:
                text += txt + '\n'
    print(text)
    
    if ix == 5:
        break
    
    print('\n\n=======================\n')
    
```
